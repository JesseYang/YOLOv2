```python
import os
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
file = "cmdt_train.txt"
```

```python
with open(file) as f:
    content = f.readlines()
```

```python
record = content[15]
```

```python
record = record.replace('\n', '')
elements = record.split(' ')
```

```python
print(elements[0])

img = cv2.imread(elements[0])
ori_height, ori_width, _ = img.shape
height = 416
width = 416
height_rate = height / ori_height
width_rate = width / ori_width
img = cv2.resize(img, (height, width))
elements[1:] = [float(num) for num in elements[1:]]
i = 1
while i < len(elements):
    xmin = elements[i] * width_rate
    ymin = elements[i + 1] * height_rate
    xmax = elements[i + 2] * width_rate
    ymax = elements[i + 3] * height_rate

    cv2.rectangle(img,
                  (int(xmin), int(ymin)),
                  (int(xmax), int(ymax)),
                  (0, 0, 255),
                  3)
    i += 5
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
