```python
import os
import xml.etree.ElementTree as ET 
import struct
import numpy as np
import cv2
import random
from matplotlib import pyplot as plt
%matplotlib inline
from IPython.core.debugger import Tracer
import shutil
```

```python
ROOT_DIR = 'doodle_detection/OwlDetection'
files = os.listdir(ROOT_DIR)
```

```python
files = [file for file in files if not file.endswith('.xml')]
# print(files)
file_name = open("draw_all.txt", "w")
for record in files:
    file_paths = os.path.join(ROOT_DIR, record)
    imgs = os.listdir(os.path.join(ROOT_DIR, record))
    for path in imgs:
        img_file_paths = os.path.join(file_paths, path)
        file_name.write(img_file_paths + '\n')
file_name.close()    
```
